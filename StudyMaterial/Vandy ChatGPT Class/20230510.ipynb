```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for the plot
x = np.linspace(-10, 10, 100)
y = 2*x + 5

# Plot the data
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear equation: y = 2x + 5')
plt.grid()

# Calculate when the graph hits the y-intercept
y_intercept = 0
x_intercept = (y_intercept - 5) / 2
print('The graph hits the y-intercept at x =', x_intercept)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x = np.linspace(-5, 5, 100)
y = 2 * x**2 + 3 * x + 1

# Plot the data and trend line
plt.scatter(x, y)
plt.plot(x, y, color='red')

# Calculate the y-intercept
a, b, c = np.polyfit(x, y, 2)  # Quadratic fit
y_intercept = c

# Print the y-intercept
print("Y-intercept:", y_intercept)

# Show the plot
plt.show()
```

```python
import random

slime = {'HP':10, 'minAttack':1, 'maxAttack':2}
shadow = {'HP':12, 'minAttack':2, 'maxAttack':5}
heroHP = 10

slimeHP = 10

heroName = ''

heroName = input('What is your name, hero?\n')

battleStatus = 'Engaged'
def battle(heroHP, slimeHP,battleStatus):
    while heroHP > 0 and slimeHP > 0 and battleStatus == 'Engaged':

        print('Choose your action:\n')
        print('1: Fight\n')
        print('2: Flee\n')
        action = int(input('Action:'))
        
        heroDamage = random.randint(1,3)
        slimeDamage = random.randint(slime['minAttack'],slime['maxAttack'])

        if action == 1:
        
            slimeHP = slimeHP - heroDamage
            print('The slime took ' + str(heroDamage) + ' damage.\n')
            if slimeHP <= 0:
                print('You have successfully defeated the slime! EXP gained: 10 point!')
                break
            heroHP = heroHP - slimeDamage
            print(heroName + ' took ' + str(slimeDamage) + ' damage.\n')
            print('You have ' + str(heroHP) + ' HP left.\n')
            if heroHP <= 0:
                print('You have dead! The world has plunged into forever darkness!')
                break
        else :
            if random.randint(1,2) == 2:
                battleStatus = 'Inactive'
                print('Successfully escaped!\n')
                break
            else:
                print('Unable to escape!\n')
                heroHP = heroHP - slimeDamage
                print(heroName + ' took ' + str(slimeDamage) + ' damage.\n')
                print('You have ' + str(heroHP) + ' HP left.\n')
                if heroHP <= 0:
                    print('You have dead! The world has plunged into forever darkness!')
                    break
```

```python
battle(heroHP, slime['HP'],'Engaged')
```

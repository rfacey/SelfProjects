```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix
from pytrends.request import TrendReq
import time
pytrend = TrendReq(hl='en-US', tz=360)
```

```python
#import PokemonData using pandas
pokemonData=pd.read_csv('PokemonData.csv', header = 0)

pokemonDataUnmodified = []

with open("PokemonData.csv", "r") as data_file:
    for row in data_file:
        tempRow = row.strip()
        tempRow = tempRow.split(",")
        pokemonDataUnmodified.append(list(tempRow))
```

```python
pokemonDataUnmodified[0:9]
```

```python
#delete row with headers

del pokemonDataUnmodified[0]
pokemonDataUnmodified[0:9]
```

```python
#remove columns that will not be used. name isn't needed and Ndex is a primary key (more or less)
##pokemonData = pokemonData.drop(['Pokémon'], axis=1) 
pokemonData = pokemonData.drop(['Ndex'], axis=1) 

#change data in each column to include what column they are in
#this will be VERY useful later on
pokemonData['Type1'] = 'Type1_' + pokemonData['Type1']
pokemonData['Type2'] = 'Type2_' + pokemonData['Type2']
pokemonData['Variance'] = 'Variance_' + pokemonData['Variance']
pokemonData['Generation'] = 'Generation_' + pokemonData['Generation'].astype(str)
pokemonData['Evolution'] = 'Evolution_' + pokemonData['Evolution'].astype(str)
pokemonData['Rarity'] = 'Rarity_' + pokemonData['Rarity']

#change the NewSnap column to 0/1, only include initially included in game (no DLC)
def newSnapCheck(data):
  if data['NewSnap'] == 1:
    return 1
  else:
    return 0

pokemonData['NewSnap'] = pokemonData.apply(newSnapCheck, axis=1)
```

```python
dataset = []

keyword = ['Bulbasaur']
pytrend.build_payload(
    kw_list=keyword,
    timeframe='2018-04-30 2021-04-30',
    geo ='US',
    cat=0
    )
data = pytrend.interest_over_time()
if not data.empty:
    data = data.drop(labels=['isPartial'],axis='columns')
    dataset.append(data)

result = pd.concat(dataset, axis=1)
```

```python
#do one hot encoding (aka dummy variables in pandas) on categorical columns

type1_occurance = pokemonData['Type1']
type1_occurance = pd.Series(type1_occurance)
type1_dummy = pd.get_dummies(type1_occurance)

type2_occurance = pokemonData['Type2']
type2_occurance = pd.Series(type2_occurance)
type2_dummy = pd.get_dummies(type2_occurance)

variance_occurance = pokemonData['Variance']
variance_occurance = pd.Series(variance_occurance)
variance_dummy = pd.get_dummies(variance_occurance)

generation_occurance = pokemonData['Generation']
generation_occurance = pd.Series(generation_occurance)
generation_dummy = pd.get_dummies(generation_occurance)

evolution_occurance = pokemonData['Evolution']
evolution_occurance = pd.Series(evolution_occurance)
evolution_dummy = pd.get_dummies(evolution_occurance)

rarity_occurance = pokemonData['Rarity']
rarity_occurance = pd.Series(rarity_occurance)
rarity_dummy = pd.get_dummies(rarity_occurance)
```

```python
#attach the dummy variables
pokemonData = pd.concat([pokemonData, type1_dummy], axis=1)
pokemonData = pd.concat([pokemonData, type2_dummy], axis=1)
pokemonData = pd.concat([pokemonData, variance_dummy], axis=1)
pokemonData = pd.concat([pokemonData, generation_dummy], axis=1)
pokemonData = pd.concat([pokemonData, evolution_dummy], axis=1)
pokemonData = pd.concat([pokemonData, rarity_dummy], axis=1)

#remove the categorical columns
pokemonData = pokemonData.drop(['Type1'], axis=1) 
pokemonData = pokemonData.drop(['Type2'], axis=1) 
pokemonData = pokemonData.drop(['Variance'], axis=1) 
pokemonData = pokemonData.drop(['Generation'], axis=1) 
pokemonData = pokemonData.drop(['Evolution'], axis=1) 
pokemonData = pokemonData.drop(['Rarity'], axis=1) 
```

```python
pokemonData.head(9)
```

```python
#separate the NewSnap column to test for it
y = pokemonData['NewSnap']
X = pokemonData.drop(['NewSnap'], axis=1)
```

```python
#split the data for training/testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 25)
```

```python
#check to make sure the columns look clean in training
X_train.info()
```

```python
#check to make sure the columns look clean in testing
X_test.info()
```

```python
#do a decision tree classifier based on the training sets
classifier=DecisionTreeClassifier()
classifier=classifier.fit(X_train,y_train)
```

```python
#use the created prediction model on the testing set
predictions = classifier.predict(X_test)
```

```python
#check results of the decision tree
print(classification_report(y_test,predictions))
```

```python
#get confusion matrix for counts
print(confusion_matrix(y_test,predictions))
```

```python
#find importance based off classifier
dt_fi = pd.DataFrame(classifier.feature_importances_)
names = pd.DataFrame(list(X.columns))
df_feat_imp = pd.concat([dt_fi, names], axis = 1)

df_feat_imp.columns = ['Importance', 'Features']
df_feat_imp.sort_values('Importance', ascending = False)
```

```python
featureCount = [[],[],[],[],[]]

Evolution_1_count = 0
Generation_1_count = 0
Evolution_3_count = 0
Evolution_2_count = 0
Generation_2_count = 0

TotalIncluded = 0

Generation_3_count = 0
Generation_4_count = 0
Generation_5_count = 0
Generation_6_count = 0
Generation_7_count = 0
Generation_8_count = 0

for row in range(0, len(pokemonDataUnmodified)):
    if pokemonDataUnmodified[row][6] == '1' \
        and pokemonDataUnmodified[row][8] == '1':
        Evolution_1_count = Evolution_1_count + 1
    if pokemonDataUnmodified[row][5] == '1' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_1_count = Generation_1_count + 1
    if pokemonDataUnmodified[row][6] == '3' \
        and pokemonDataUnmodified[row][8] == '1':
        Evolution_3_count = Evolution_3_count + 1
    if pokemonDataUnmodified[row][6] == '2' \
        and pokemonDataUnmodified[row][8] == '1':
        Evolution_2_count = Evolution_2_count + 1
    if pokemonDataUnmodified[row][5] == '2' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_2_count = Generation_2_count + 1
    
    if pokemonDataUnmodified[row][8] == '1':
        TotalIncluded = TotalIncluded + 1
    
    if pokemonDataUnmodified[row][5] == '3' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_3_count = Generation_3_count + 1
    if pokemonDataUnmodified[row][5] == '4' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_4_count = Generation_4_count + 1
    if pokemonDataUnmodified[row][5] == '5' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_5_count = Generation_5_count + 1
    if pokemonDataUnmodified[row][5] == '6' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_6_count = Generation_6_count + 1
    if pokemonDataUnmodified[row][5] == '7' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_7_count = Generation_7_count + 1
    if pokemonDataUnmodified[row][5] == '8' \
        and pokemonDataUnmodified[row][8] == '1':
        Generation_8_count = Generation_8_count + 1

featureNames = ['Evolution 1','Generation 1','Evolution 3','Evolution 2','Generation 2']
featureCount = [Evolution_1_count,Generation_1_count,Evolution_3_count,Evolution_2_count,Generation_2_count]
```

```python
print(featureNames[0] + ' Count: ' + str(featureCount[0]))
print(featureNames[1] + ' Count: ' + str(featureCount[1]))
print(featureNames[2] + ' Count: ' + str(featureCount[2]))
print(featureNames[3] + ' Count: ' + str(featureCount[3]))
print(featureNames[4] + ' Count: ' + str(featureCount[4]))

print('\n')

print('Total Included: ' + str(TotalIncluded))

print('\n')

print('Generation 3 Count: ' + str(Generation_3_count))
print('Generation 4 Count: ' + str(Generation_4_count))
print('Generation 5 Count: ' + str(Generation_5_count))
print('Generation 6 Count: ' + str(Generation_6_count))
print('Generation 7 Count: ' + str(Generation_7_count))
print('Generation 8 Count: ' + str(Generation_8_count))
```

```python
#https://people.duke.edu/~ccc14/pcfb/numpympl/MatplotlibBarPlots.html
#https://stackoverflow.com/questions/4804005/matplotlib-figure-facecolor-background-color
    
fig = plt.figure()
fig, ([ax1, ax2]) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
fig.canvas.set_window_title('A Boxplot Example')
fig.subplots_adjust(left=0.075, right=0.95, top=0.9, bottom=0.25)
fig.set_facecolor('white')

fig.set_facecolor('white')

## the data
N = len(featureNames)

## necessary variables
ind = np.arange(N)                # the x locations for the groups
width = 0.35                      # the width of the bars

## the bars
rects1 = ax1.bar(featureNames, featureCount, width,
                color='steelblue')


# axes and labels
ax1.set_xlim(-(width*1.5),len(ind)-(width*1.5))
ax1.set_ylim(0,80)
ax1.set_ylabel('Pokemon With Trait', fontsize=12)
ax1.set_title('Count Of Features')
ax1.set_facecolor('white')
xTickMarks = featureNames
ax1.set_xticks(ind)
xtickNames = ax1.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=0, fontsize=10)


############

#https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html
# Pie chart, where the slices will be ordered and plotted counter-clockwise:

pieFeatures = ['Generation 1','Generation 2','Generation 3','Generation 4','Generation 5' \
               ,'Generation 6','Generation 7','Generation 8']
pieValues = [Generation_1_count,Generation_2_count,Generation_3_count,Generation_4_count,Generation_5_count \
             ,Generation_6_count,Generation_7_count,Generation_8_count]


rects2 = ax2.pie(pieValues, labels=pieFeatures, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax2.set_title('Pokemon By Generation')

plt.show()
# plt.savefig("featureBarChart.png", facecolor='white')
```

## Artificial Intelligence vs Neural Networks vs Machine Learning

- **Artificial Intelligence**: The effort to automate intellectual tasks normally performed by humans. It can be simple or complex. 

- **Machine Learning**: Part of A.I. Machine learning figures out the rules for us. The goal for machine learning models is to have the highest accuracy possible.



- Machine Learning Summary: Rather than giving the program the rules, an algorithm finds the rules for us.
- **Neural Networks**: A form of machine learning that uses a layered representation of data. In "deep learning", we have more than two layers. Also called a "multi stage information extraction process". Not actually modeled after the brain. 



- **Features**: Information given to the model to get some output. The input information.
- **Label**: What we are trying to predict. The output information.

```python
#https://stackoverflow.com/questions/16627227/problem-http-error-403-in-python-3-web-scraping

from urllib.request import Request, urlopen

req = Request(
    url='https://finance.yahoo.com/u/yahoo-finance/watchlists/biotech-and-drug-stocks/', 
    headers={'User-Agent': 'Mozilla/5.0'}
)
webpage = urlopen(req).read()
```

```python
# Create byte
b = webpage

# Using the decode() method
s = b.decode('utf-8')

# Using the str() constructor
s = str(b)

# Using the bytes.decode() method
s = bytes.decode(b, 'utf-8')

# Using the bytearray.decode() method
s = b.decode('utf-8')

# Using the codecs.decode() method
import codecs
string = codecs.decode(b, 'utf-8')
```

```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("Encountered a start tag:", tag)

    def handle_endtag(self, tag):
        print("Encountered an end tag :", tag)

    def handle_data(self, data):
        print("Encountered some data  :", data)

parser = MyHTMLParser()
parser.feed(string)
```

```python
tag = "tbody"
tagLength = len(tag)

locationStart = 0
locationEnd = 0

tagStartFound = 0
tagEndFound = 0

tempString = ""


for i in range(0, len(string)):
    tempString = tempString + string[i]
    if len(tempString) > tagLength:
        tempString = tempString[1:tagLength+1]
    if tempString == tag:
        if tagStartFound == 1 and tagEndFound == 0:
            locationEnd = i
            tagEndFound = 1
        if tagStartFound == 0:
            locationStart = i
            tagStartFound = 1
    if (tagStartFound == 1) and (tagEndFound ==1):
        break
print(locationStart)
print(locationEnd)

cutString = string[locationStart-tagLength:locationEnd+2]
```
